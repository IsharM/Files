```python
plt.figure(1,figsize=(25, 80), dpi=180)
figIndex =1
dict_State = []
for column in GSDP_growth_df:
    if column != 'Duration':
        plt.subplot(10,3,figIndex)
        sns.scatterplot(x='Duration', y=column, data= GSDP_growth_df)
        figIndex+=1
        range_len = len(GSDP_growth_df[column].dropna())
        m, c = np.polyfit(range(range_len), GSDP_growth_df[column].dropna(),1)
        dict_State.append([column, m])
        plt.plot(range(range_len),m*range(range_len)+c,'r',label=round(m,1))
        plt.legend(loc='upper right')
```

```python
State_Performance = pd.DataFrame(dict_State,columns=['States','Slope'])
State_Performance.sort_values(["Slope"], axis=0, ascending=False, inplace=True)
State_Performance
```
